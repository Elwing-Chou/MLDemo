```python
import glob
import pandas as pd
import os
df = pd.DataFrame()
for dn in glob.glob("chinese_news_trans/*"):
    basedn = os.path.basename(dn)
    for fn in glob.glob(dn + "/*.txt"):
        with open(fn, "r", encoding="utf-8") as f:
            news = f.read()
        row = {"content":news, "category":basedn}
        df = df.append(row, ignore_index=True)
df
```
