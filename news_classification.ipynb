```python
import glob
import pandas as pd
import os
train_df = pd.DataFrame()
for dn in glob.glob("chinese_news_trans/*"):
    basedn = os.path.basename(dn)
    for fn in glob.glob(dn + "/*.[tT][xX][tT]"):
        with open(fn, "r", encoding="utf-8") as f:
            news = f.read()
        row = {"content":news, "category":basedn}
        train_df = train_df.append(row, ignore_index=True)
train_df
```

```python
import glob
import pandas as pd
import os
test_df = pd.DataFrame()
for dn in glob.glob("chinese_news_test/*"):
    basedn = os.path.basename(dn)
    for fn in glob.glob(dn + "/*.[tT][xX][tT]"):
        with open(fn, "r", encoding="utf-8") as f:
            news = f.read()
        row = {"content":news, "category":basedn}
        test_df = test_df.append(row, ignore_index=True)
test_df
```

```python
cat = train_df["category"].unique()
trans = {v:i for i, v  in enumerate(cat)}
y_train = train_df["category"].replace(trans)
y_test = test_df["category"].replace(trans)
y_test
```

```python
from jieba import cut, set_dictionary
from urllib.request import urlretrieve
import os
if not os.path.exists("dict.txt.big"):
    print("幫下載大辭典")
    url = "https://github.com/fxsjy/jieba/raw/master/extra_dict/dict.txt.big"
    urlretrieve(url, "dict.txt.big")
    
set_dictionary("dict.txt.big")
# Series.apply(func)
def newscut(s):
    s = (s.replace("\r", " ")
      .replace("\n", " "))
    return " ".join(cut(s))
x_train = train_df["content"].apply(newscut)
x_test = test_df["content"].apply(newscut)
x_test
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vec = CountVectorizer()
x_train_vec = vec.fit_transform(x_train)
x_test_vec = vec.transform(x_test)
print(x_test_vec)
```

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
clf.fit(x_train_vec, y_train)
```

```python
from sklearn.metrics import accuracy_score
pre = clf.predict(x_test_vec)
accuracy_score(pre, y_test)
```

```python
from sklearn.metrics import confusion_matrix
labels = train_df["category"].unique()
pd.DataFrame(confusion_matrix(y_test, pre),
             columns=[(l + "(預測)") for l in labels],
             index=[(l + "(正確)") for l in labels])
```

```python
p = input("給篇新聞!")
p = newscut(p)
p_vec = vec.transform([p])
print("應該是哪個分類?", labels[clf.predict(p_vec)[0]])
```
